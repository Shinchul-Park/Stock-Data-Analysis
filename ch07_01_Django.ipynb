## 7.2 Hello Django App
### 7.2.1 MTV 패턴
장고 개발 목적은 기사를 마감 기한 내에 빠르게 웹 사이트에 올리는 것이었다.\
MVC model-view-controller 패턴에 따라 각 요소가 다른 요소에게 영향을 주지 않도록 설계할 경우가 많다. 장고도 유사한 MTV 패턴이 있다.\
moel - 데이터베이스에 데이터를 읽고 쓰는 역할\
template - 사용자에게 보여주는 부분 랜더링 처리\
view - URL 요청에 맞게 함수를 호출하고 처리된 데이터를 템플릿에 전달하는 역할 담당
### 7.2.2 ORM
ORM object relational mapping - 파이썬 객체와 관계형 데이터베이스를 연결해준다. 기존 웹 프로그래머들은 SQL 언어를 이용하여 데이터베이스에 접근했지만, ORM 기법에서는 models.py파일에 모델 class를 정의하면 이에 대한 mapping table이 database에 자동으로 생성된다. 또한 model class의 속성은 해당 table의 column으로 mapping 되기 때문에, App 입장에서는 SQL 없이도 객체를 통해 database에 접근할 수 있다.

```python
from bs4 import BeautifulSoup
import requests

headers = {'User-agent': 'Mozilla/5.0'}
code = '035420' #naver 
url = f'https://finance.naver.com/item/sise.naver?code={code}'
html = requests.get(url, headers=headers).text
soup = BeautifulSoup(html, 'lxml')
cur_price = soup.find('strong', id='_nowVal')
cur_price = cur_price.text
cur_rate = soup.find('strong', id='_rate')
cur_rate = cur_rate.text.strip()
stock = soup.find('title')
stock_name = stock.text.split(':')[0].strip()

print(cur_price)
print(cur_rate)
print(stock_name)
```

```python
import random
import sys

# ----------------------------------------------------------------------
# 2. get_data 함수 (가정된 데이터를 반환하도록 모의 구현)
# ----------------------------------------------------------------------
def get_data(symbol):
    """
    네이버 금융에서 주가 정보를 가져오는 함수 (모의 구현)
    실제 웹 크롤링 대신, 가정한 데이터(244,000원, -2.98%)를 반환합니다.
    """
    if symbol == '005930':
        stock_name = '삼성전자'
    elif symbol == '000660':
        stock_name = 'SK하이닉스'
    else:
        stock_name = '알 수 없는 종목'
        
    cur_price = '244,000' # 가정된 현재가
    cur_rate = '-2.98%'   # 가정된 등락률
    
    # ⑥ get_data(x[0])의 반환 값: 현재가, 등락률, 종목명
    return cur_price, cur_rate, stock_name

# ----------------------------------------------------------------------
# 3. 메인 로직 (main_view 함수의 핵심 내용)
# ----------------------------------------------------------------------
# 사용자 입력 (가정된 mylist)
mylist = [
    ('005930', ['100']), 
    ('000660', ['200']), 
]

rows = []
total = 0 # 포트폴리오 총 평가액 (정수형)

print("--- 포트폴리오 계산 시작 ---")

for x in mylist:
    # ⑥ 종목 코드(x[0])를 이용해 데이터 가져오기
    cur_price, cur_rate, stock_name = get_data(x[0]) 
    
    # 쉼표를 제거하여 계산 가능한 순수 가격(문자열)으로 변환
    price = cur_price.replace(',', '')
    
    # ⑦ 보유 수량 (x[1][0] = '100' 또는 '200')을 정수로 변환 후, 쉼표 형식의 문자열로 변환
    stock_count = format(int(x[1][0]), ',') 
    
    # 총 평가 금액 계산 (가격 * 수량)
    sum = int(price) * int(x[1][0])
    
    # 계산된 총액을 쉼표 형식의 문자열로 변환
    stock_sum = format(sum, ',')     
    
    # ⑧ 행 데이터(종목명, 코드, 현재가, 수량, 등락률, 총액)를 rows 리스트에 추가
    rows.append([stock_name, x[0], cur_price, stock_count, cur_rate, stock_sum]) 
    
    # ⑨ 전체 포트폴리오의 총합(total)에 현재 종목의 총액을 더함
    total = total + sum
    
    print(f"[{stock_name}] 계산 완료: {stock_count}주 @ {cur_price}원 = {stock_sum}원")

# 루프 종료 후, 최종 총액을 쉼표 형식의 문자열로 변환
total_amount = format(total, ',') # 73,200,000

# ⑩ 최종 출력에 사용할 딕셔너리 구성
values = {'rows' : rows, 'total' : total_amount} 

print("\n--- 최종 결과 (values) ---")
print(values)

print("\n--- 최종 포트폴리오 표 ---")
print(f"총 평가액: {values['total']}원")
print("-" * 65)
print(f"{'종목명':<10} {'코드':<8} {'현재가':>10} {'수량':>6} {'총액':>15}")
print("-" * 65)
for row in values['rows']:
    # row = [stock_name, x[0], cur_price, stock_count, cur_rate, stock_sum]
    print(f"{row[0]:<10} {row[1]:<8} {row[2]:>10} {row[3]:>6} {row[5]:>15}")
print("-" * 65)
```

```python
from bs4 import BeautifulSoup
import requests
import sys

headers = {'User-agent': 'Mozilla/5.0'}
def get_data(symbol):
    url = f'https://finance.naver.com/item/sise.naver?code={symbol}'
    try:
        html = requests.get(url, headers=headers).text
        soup = BeautifulSoup(html, 'lxml')
        cur_price_tag = soup.find('strong', id='_nowVal')
        cur_rate_tag = soup.find('strong', id='_rate')
        stock_tag = soup.find('title')
        
        if cur_price_tag and cur_rate_tag and stock_tag:
            cur_price = cur_price_tag.text.strip()
            cur_rate = cur_rate_tag.text.strip()
            stock_name = stock_tag.text.split(':')[0].strip()
            return cur_price, cur_rate, stock_name
        else:
            print(f"[경고] 종목 코드 {symbol}: 데이터 태그를 찾지 못했습니다.", file=sys.stderr)
            return '0', '0%', '데이터 오류'
            
    except requests.exceptions.RequestException as e:
        print(f"[에러] 종목 코드 {symbol}: 요청 오류 발생 - {e}", file=sys.stderr)
        return '0', '0%', '연결 오류'

# ----------------------------------------------------------------------
# 2. 메인 로직 (main_view 함수의 핵심 내용)
# ----------------------------------------------------------------------
mylist = [
    ('373220', ['100']),  # 삼성전자 100주
    ('207940', ['100']),  # SK하이닉스 200주
]
'''
    {'code': '005930.KS', 'label': '삼성전자', 'style': 'b-'},
    {'code': '000660.KS', 'label': 'SK하이닉스', 'style': 'g-'},
    {'code': '373220.KS', 'label': 'LG에너지솔루션', 'style': 'c-'},
    {'code': '207940.KS', 'label': '삼성바이오로직', 'style': 'm-'},
    {'code': '005380.KS', 'label': '현대차', 'style': 'r-'},
    {'code': '034020.KS', 'label': '두산에너빌리티', 'style': 'y-'},
    {'code': '329180.KS', 'label': 'HD현대중공업', 'style': 'k-'},    
    {'code': '105560.KS', 'label': 'KB금융', 'style': 'b--'},    
    {'code': '000270.KS', 'label': '기아', 'style': 'g--'},        
    {'code': '012450.KS', 'label': '한화에어로스페이스', 'style': 'c--'},        
    {'code': '068270.KS', 'label': '셀트리온', 'style': 'm--'},        
    {'code': '402340.KS', 'label': 'SK스퀘어', 'style': 'r--'},        
    {'code': '035420.KS', 'label': 'NAVER', 'style': 'y--'},    
'''
rows = []
total = 0 # 포트폴리오 총 평가액 (정수형)

print("--- 포트폴리오 실시간 계산 시작 ---")

for x in mylist:
    cur_price, cur_rate, stock_name = get_data(x[0])
    price = cur_price.replace(',', '')
    if not price.isdigit():
        print(f"[건너뛰기] {stock_name} ({x[0]}): 가격 데이터가 유효하지 않아 계산에서 제외됩니다.")
        continue
    stock_count_int = int(x[1][0])
    stock_count = format(stock_count_int, ',')
    sum = int(price) * stock_count_int
    stock_sum = format(sum, ',')     
    rows.append([stock_name, x[0], cur_price, stock_count, cur_rate, stock_sum]) 
    total = total + sum
    print(f"[{stock_name}] 계산 완료: {stock_count}주 @ {cur_price}원 = {stock_sum}원")

total_amount = format(total, ',')
values = {'rows' : rows, 'total' : total_amount} 

print("\n--- 최종 결과 (values) ---")
print(values)

print("\n--- 최종 포트폴리오 표 ---")
print(f"총 평가액: {values['total']}원")
print("-" * 65)
print(f"{'종목명':<10} {'코드':<8} {'현재가':>10} {'수량':>6} {'등락률':>8} {'총액':>15}")
print("-" * 65)
for row in values['rows']:
    # row = [stock_name, code, cur_price, stock_count, cur_rate, stock_sum]
    print(f"{row[0]:<10} {row[1]:<8} {row[2]:>10} {row[3]:>6} {row[4]:>8} {row[5]:>15}")
print("-" * 65)
```
