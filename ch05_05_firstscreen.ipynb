### 6.5.1 시장지표
시장지표 Market Indicator, 시장의 흐름을 나타내는 지표를 추세 trend 지표(이동평균, 이동평균 수렴확산 MACD(Moving Average Convergence Divergence) )\
시장이 횡보할 때 잘못된 신호를 보낼 수 있다. \
스토캐스틱 Stochastic 이나 RSI(Relative Strength Index)처럼 과거 일정 기간의 가격 범위 않에서 현재 가격의 상대적인 위치는 나타내는 지표를 Oscillator라고 함.\
오실레이터는 횡보장에서 전환점을 포착하는 데 적합하지만 가격보다 앞서 변하는 경향이 있다.

### 6.5.2 단순 이동평균
simple moving average, SMA
$$단순\ 이동평균\ SMA=\frac{P_1+P_2+P_3+P_N}{N} $$
* P: 가격
* N: 이동평균이 구해지는 시간
단순이동평균은 오래된 가격의 변동과 최근 가격의 변동을 동일하게 반영하기 때문에 최근 가격의 변동이 왜곡될 가능성이 있다.

### 6.5.3 지수 이동평균
exponential moving average, EMA, 최근 data에 가중치를 부여해 단순 이동평균에 비해서 최근의 데이터 변동을 잘 반영하도록 설계됨.
$$지수\ 이동평균\ EMA=P_{today} * K + EMA_{yesterday} *(1-K) $$
* K: $\frac{2}{N+1} $
* N: 지수 이동평균 일수
* $P_{today}$: 오늘의 가격
* $EMA_{yesterday}$: 어제의 지수 이동평균
최근 거래일에 더 많은 가중치를 주므로 최근 가격의 변동을 더 잘 반영, 오래된 지수 이동평균 데이터가 천천히 사라지므로 오래된 데이터가 빠져날 때 지수 이동평균이 금등락하지 않는다.\
알렉산터 엘더에 따르면 이동평균의 기간은 시장 the dominant market  사이클의 절반 정도, 즉 20일 주기를 발견했다면 10일 이동평균선을 사용. 그는 매매할 때 13일 지수이동평균을 사용.

### 6.5.4 이동평균 수렴확산(MACD)
제럴드 아펠 Gerald Appel, 세가지 지수 이동평균선을 이용해 이동평균 수렴확산(moving average convergence divergence, MACD) 개발\
MACD 선은 종가의 12일 지수 이동평균선, 26일 지수 이동평균선을 뺀 것, 가격 변화에 상대적으로 빨리 반응, 신호선은 MACD 선의 9일 지수 이동평균을 구한 선, MACD선을 평활화시킨 것이기 때문에 가격 변화에 상대적으로 늦게 반응한다.\
빠른 MACD선이 늦은 신호선을 상향 돌파하는 것은 매수세가 시장을 주도한다는 뜻이므로 매수적 관점에서 대응, 반대로 빠른 MACD선이 늦은 신호선을 하향 돌파할 때는 매도 관점에서 대응.

### 6.5.5 MACD 히스토그램
MACD histogram은 원래의 MACD 보다 매수와 매도 상태를 더 잘 표현, 단순히 매수와 매도의 비중을 표시할 뿐만 아니라, 강해지고 있는지 약해지고 있는지를 보여줌.
$$MACD\ histogram=MACD_{line}-MACD_{signal} $$
MACD historgam 기울기는 histogram 중심선 위에 있는지 아래에 있는지 보다 중요, 현재 봉이 이전 봉보다 높다면 기울기는 올라가고 있으므로 매수, 최고의 매수 신호는 MACD histogram이 중심선 아래에 있고, 기울기가 상향 반전하고 있을 때 발생한다.

### 6.5.6 스토캐스틱
Stochastic 조지 레인 George Lane, oscilator, 지난 n일 동안의 거래 범위에서 현재 가격 위치를 백분율로 나타냄.\
14일 스토캐스틱이 70이면 지난 14일간 거래 범위에서 최저점과 최고점 사이 70%에 위치해 있다는 의미. 일반적으로 80이상은 과매수, 20이하는 과매도 상태\
빠른 선 %K, 느린 선 %D, %K의 기간은 14일로 설정, 알렉산더 엘더는 짧은 반전을 잡아내기 용이한 5일로 설정한다. 반면 기간이 길면 중요 변곡점을 잡아내는 데 유용.
$$\% K=\frac{C_{today}-L_n}{H_n-Ln}*100 $$
* $C_{today}$: 오늘의 종가, $H_n$: 선정된 기간의 고점, $L_n$: 선정된 기간의 저점, $n$: 트레이더에 의해 선정된 기간
$$\% D=\frac{(C_{today}-L_n)의 3일간\ 합계}{(H_n-L_n)의 3일간\ 합계}*100 $$
스토캐스틱은 시장이 박스권에서 움직일 때는 잘 작동하지만 시장이 추세에 들어갈 때는 그렇지 않다.\
스토캐스틱은 장기 추세 추종형 지표와 결합해서 사용해야 한다.

```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import matplotlib.dates as mdates
import mplfinance as mpf
from mplfinance.original_flavor import candlestick_ohlc 
#import sys
#import os

investar_parent_path = 'C:\\myPackage'
if investar_parent_path not in sys.path:
    sys.path.append(investar_parent_path)

from Investar.Analyzer import MarketDB
mk = MarketDB()
df = mk.get_daily_price('엔씨소프트', '2025-06-02')
print(df.tail())

# MACD 지표 계산
ema60 = df.close.ewm(span=60).mean()
ema130 = df.close.ewm(span=130).mean()
macd = ema60 - ema130
signal = macd.ewm(span=45).mean()
macdhist = macd - signal

df = df.assign(ema130=ema130, ema60=ema60, macd=macd, signal=signal, macdhist=macdhist).dropna()
df['number'] = df.index.map(mdates.date2num)
ohlc = df[['number','open','high','low','close']]

# 차트 그리기
plt.figure(figsize=(9, 6))
p1 = plt.subplot(2, 1, 1)
plt.title('Triple Screen Trading - First Screen (Samsung)')
plt.grid(True)
candlestick_ohlc(p1, ohlc.values, width=.6, colorup='red', colordown='blue')
p1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.plot(df.number, df['ema130'], color='c', label='EMA130')
plt.legend(loc='best')

p2 = plt.subplot(2, 1, 2)
plt.grid(True)
p2.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.bar(df.number, df['macdhist'], color='m', label='MACD-Hist')
plt.plot(df.number, df['macd'], color='b', label='MACD')
plt.plot(df.number, df['signal'], 'g--', label='MACD-Signal')
plt.legend(loc='best')

plt.show()
```

## 6.6.2 두번째 창 - 사장 파도
second screen: 첫번째 창의 추세 방향과 역행하는 파도 market wave를 파악하는 데 오실레이터를 활용한다. 오실레이터는 시장이 하락할 때 매수 기회를 제공하고, 시장이 상승할 때 매도 기회를 제공한다. 주봉 추세가 상승하고 있을 때 일봉 추세가 하락하면 매수 기회로 본다.\
**주간 히스토그램이 상승하고 있을 때 스토캐스틱이 30 아래로 내려가면 매수하고, 주간 MACD 히스토그램이 하락하고 있을 때 스토캐스틱이 70 위로 올라가면 매도하라고 나와 있다. 주간 히스토그램보다 130일 지수 이동평균이 신뢰성이 높기 때문에, 130일 지수 이동평균과 스토캐스틱을 함께 사용한다.**


```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import matplotlib.dates as mdates
import mplfinance as mpf
from mplfinance.original_flavor import candlestick_ohlc 
#import sys
#import os

investar_parent_path = 'C:\\myPackage'
if investar_parent_path not in sys.path:
    sys.path.append(investar_parent_path)

from Investar.Analyzer import MarketDB
mk = MarketDB()
df = mk.get_daily_price('엔씨소프트', '2025-05-02')
print(df.tail())

ema60 = df.close.ewm(span=60).mean()
ema130 = df.close.ewm(span=130).mean() 
macd = ema60 - ema130
signal = macd.ewm(span=45).mean() 
macdhist = macd - signal

df = df.assign(ema130=ema130, ema60=ema60, macd=macd, signal=signal, macdhist=macdhist).dropna()
df['number'] = df.index.map(mdates.date2num)
ohlc = df[['number','open','high','low','close']]

ndays_high = df.high.rolling(window=14, min_periods=1).max()      # ①
ndays_low = df.low.rolling(window=14, min_periods=1).min()        # ②
fast_k = (df.close - ndays_low) / (ndays_high - ndays_low) * 100  # ③
slow_d= fast_k.rolling(window=3).mean()                           # ④
df = df.assign(fast_k=fast_k, slow_d=slow_d).dropna()             # ⑤

plt.figure(figsize=(9, 7))
p1 = plt.subplot(2, 1, 1)
plt.title('Triple Screen Trading - Second Screen (NCSOFT)')
plt.grid(True)
candlestick_ohlc(p1, ohlc.values, width=.6, colorup='red', colordown='blue')
p1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.plot(df.number, df['ema130'], color='c', label='EMA130')
plt.legend(loc='best')

p1 = plt.subplot(2, 1, 2)
plt.grid(True)
p1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.plot(df.number, df['fast_k'], color='c', label='%K')
plt.plot(df.number, df['slow_d'], color='k', label='%D')
plt.yticks([0, 20, 80, 100]) # ⑥
plt.legend(loc='best')
plt.show()
```
