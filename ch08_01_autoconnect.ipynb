```python
import requests

def get_current_price(ticker):
    url = f"https://api.finance.naver.com/service/itemSummary.nhn?itemcode={ticker}"
    headers = {"User-Agent": "Mozilla/5.0"}

    try:
        data = requests.get(url, headers=headers).json()
        return data["now"]
    except:
        return None

print(get_current_price("005930"))
```

```python
# ETF 모두를 불러온다.
import json
import urllib.request

url_etf = 'https://finance.naver.com/api/sise/etfItemList.nhn' 
etf_raw_data = urllib.request.urlopen(url_etf).read().decode('CP949') 
etf_json_data = json.loads(etf_raw_data)
for each in etf_json_data['result']['etfItemList']: 
        item_name = each['itemname']
        item_code = each['itemcode']
        print ("ETF명 : %s / ETF코드 : %s "%(item_name,item_code))

url_etn = 'https://finance.naver.com/api/sise/etnItemList.nhn'
etn_raw_data = urllib.request.urlopen(url_etn).read().decode('CP949')
etn_json_data = json.loads(etn_raw_data)

for each in etn_json_data['result']['etnItemList']: 
        item_name = each['itemname']
        item_code = each['itemcode']
        print ("ETN명 : %s / ETN코드 : %s "%(item_name,item_code))
```

### 8.4.2 매수 목표가 계산
매수목표가 = 금일 시작 가격 + (어제 최고자 - 어제 최저가) * K \
K=0.5
### 8.4.3 이동 평균값 조회
변동성 돌파 전략의 매수 조건은 현재가가 시가+전일변동폭의 K를 돌파할 때다. 여기에 현재가가 5일 이동 평균선과 10일 이동 평균선 위에 있어야 한다.\
### 8.4.4 주식 매매 주문 방식
매수자를 위한 최우선 지정가 주문: 매수자 - 가장 높은 매수호가, 매도자 - 가장 낮은 매도호가\
매도자을 위한 최유리 지정가 주문: 매수자 - 가장 낮은 매도호가, 매도자 - 가장 높은 매수호가\
FOK(Fill or Kill): 주문 즉시 전부 체결 또는 전부 취소 조건\
IOC(Immediate or Cancel): 주문 즉시 체결 그리고 잔동 자동 취소 조건
### 8.4.5 최유리 FOK 매수 주문
오전 09:15부터 오후 03:15까지 일정 간격으로 매수 종목 후보군의 현재가, 매수 목표가, 5일 이동평균가, 10일 이동평균가를 구해서 매수 조건에 부합하는지 점검할 것이다.\
현재가가 매수 목표가, 5일 이동평균가, 10일 이동평균가보다 높을 경우 매수 조건에 부합한 것으로 보고 해당 종목을 최유리 FOK 주문 방식으로 매수한다.
### 8.4.6 최유리 IOC 매도 주문
오후 3:15부터 03:20까지 매수했던 주식 종목들을 모두 매도 청산하자. 매수한 종목은 당일 매도하는 것을 원칙.
