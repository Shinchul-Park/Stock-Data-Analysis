#### 2.1.1.5 종목별 DIV/DPS/BPS/PER/EPS 조회

| 지표 | 전체이름 | 수식 |핵심 | 의미|지표 해석 (일반적) |
|:---:|:---:|:---:|:---:|:---:| ---|
|EPS(수익성)|주당 순이익 (Earnings Per Share)|$$\text{EPS} = \frac{\text{당기순이익}}{\text{총 발행 주식 수}}$$|주식 1주당 순이익 창출 능력|높을수록 기업의 수익성이 좋음|
|PER(수익성)|주가 수익 비율 (Price Earnings Ratio)|$$\text{PER} = \frac{\text{현재 주가}}{\text{EPS}}$$|주가가 순이익 대비 몇 배인지 (가장 대중적인 수익성 평가)|낮을수록 저평가로 간주될 수 있음|
|BPS(자산가치)|주당 순자산 (Book-value Per Share)|$$\text{BPS} = \frac{\text{총 순자산 (자본)}}{\text{총 발행 주식 수}}$$|주식 1주당 순자산 가치 (회사의 장부 가치)|높을수록 기업의 재무 건전성이 좋음|
|PBR(자산가치)|주가 순자산 비율 (Price Book-value Ratio)|$$\text{PBR} = \frac{\text{현재 주가}}{\text{BPS}}$$|주가가 순자산 대비 몇 배인지 (대표적인 자산 평가)|1 미만일 경우, 장부 가치보다 주가가 낮음 (저평가 가능성)|
|DPS(배당)|주당 배당률 (Dividend Per Shared)|$$\text{DPS} = \frac{\text{총 배당금 지급액}}{\text{총 발행 주식 수}}$$|주식 1주당 주주에게 지급되는 현금 금액|꾸준히 증가, 유지, 재무상태 안정적, 주주환원의지|
|DIV(배당)|배당수익률 (Dividend Yield)|$$\text{DIV} = \frac{\text{DPS}}{\text{현재 주가}} \times 100$$|투자 원금 대비 현금 배당으로 얻는 수익률|높을수록 배당 매력이 높음|

**순자산(총자산 - 총부채)**\
PER & PBR: 둘 다 주가를 평가하는 지표이지만, $\text{PER}$은 수익 창출 능력을, $\text{PBR}$은 보유 자산 가치를 기준으로 평가합니다. 가치 투자자는 보통 $\text{PER}$과 $\text{PBR}$이 낮은 종목을 선호합니다.\
DPS & DIV: $\text{DPS}$는 배당금의 절대적인 금액을, $\text{DIV}$는 그 배당금이 주가 대비 어느 정도의 수익률을 제공하는지를 나타냅니다.\
EPS & BPS: 이 두 지표는 $\text{PER}$과 $\text{PBR}$의 분모가 되는 기업의 기초 체력 지표입니다. 이 수치가 높을수록 기업의 기초가 튼튼하다고 볼 수 있습니다.

````python
from pykrx import stock
import pandas as pd

tickers = stock.get_etf_ticker_list("20251201")

data = []

#for ticker in tickers[:5]:    # 앞에서 5개만
for ticker in tickers:    # 앞에서 5개만
    name = stock.get_etf_ticker_name(ticker)
    data.append([ticker, name])
    #print(f"{ticker} : {name}")

df = pd.DataFrame(data, columns=["ticker", "name"])
print(df)
````
